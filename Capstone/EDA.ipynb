The Affects of COVID 19 on Various Life Factors of Students

In the previous step, Data Wrangling, we removed any data that was outside of the United States in order to streamline our process. Some of the participants lived in countries that do not have states, "State" being a column of interest for us. We eliminated any missing values to ensure that our data would be as clear and concise as possible. Finally, we took a preliminary look at the data to see if we could easily identify any particular trends that would be of interest. In this next phase our analysis, Exploratory Data Analysis (or EDA), we will take an even closer look at various trends of interest and discuss some important findings.

```python
#Import packages
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
#Load the data
# Get the path to the "documents" folder
documents_folder = os.path.expanduser('~/Documents')

# Specify the file name
file_name = 'Covid_Responses.csv'

# Construct the full file path
file_path = os.path.join(documents_folder, file_name)

# Load the CSV file into a DataFrame
data = pd.read_csv(file_path)
```

Now that the data is loaded, let's begin by computing some summary statistics. This will be an initial check to understand the our dataset's numerical values. We can look at important metrics; mean, median, standard deviation, minimum, and maximum values, which will give us a better understanding of the data distribution, potential anomalies, and data quality issues like missing values or incorrect data types. These statistics inform subsequent data preprocessing steps and guide exploratory data analysis by offering initial insights into the dataset's characteristics, aiding in the formulation of hypotheses and identification of trends.

```python
# Compute summary statistics
summary_stats= data.describe()

# Display the summary statistics
print(summary_stats)
```

For good measure, let's inspect the first few rows of our dataset to ensure that the data loaded correctly and looks like what we expect.

```python
data.head()
```

Next, let's examine the distributions of age and classwork stress to gain insights into the demographic composition of the dataset and the prevalence of stress levels among students. Understanding the distribution of ages will help in identifying the age groups represented in the dataset, which can inform further analysis based on age-related trends. Similarly, exploring the distribution of classwork stress provides an overview of the prevalence and severity of stress levels among students, which is crucial for understanding the overall well-being and academic experiences of the our population.

```python
# Set up the figure and axis
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Create histograms
sns.histplot(data['Age'], ax=axes[0, 0], bins=10, kde=True)
axes[0, 0].set_title('Distribution of Age')

sns.histplot(data['Before-ClassworkStress'], ax=axes[0, 1], bins=10, kde=True)
axes[0, 1].set_title('Distribution of Classwork Stress (Before)')

sns.histplot(data['Now-ClassworkStress'], ax=axes[1, 0], bins=10, kde=True)
axes[1, 0].set_title('Distribution of Classwork Stress (Now)')

sns.histplot(data['Before-HomeworkHours'], ax=axes[1, 1], bins=10, kde=True)
axes[1, 1].set_title('Distribution of Homework Hours (Before)')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()

# Convert infinity values to NaN in the dataset
data.replace([np.inf, -np.inf], np.nan, inplace=True)

import warnings

# Suppress all FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

```

Now let's examine stress levels vs homework hours before and after the pandemic This will allows us to explore potential relationships between these variables. Is there a correlation here?

```python
# Set up the figure and axis
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Scatter plot for Before stress levels vs. Before homework hours
sns.scatterplot(x='Before-ClassworkStress', y='Before-HomeworkHours', data=data, ax=axes[0])
axes[0].set_title('Before Stress Levels vs. Before Homework Hours')

# Scatter plot for Now stress levels vs. Now homework hours
sns.scatterplot(x='Now-ClassworkStress', y='Now-HomeworkHours', data=data, ax=axes[1])
axes[1].set_title('Now Stress Levels vs. Now Homework Hours')

# Show the plots
plt.tight_layout()
plt.show()
```

Let's see if there is a correlation between stress levels, friend relationships and school environment before and after the pandemic.

To gain a more clear understanding of the correlations that exist in our dataset, let's create a heatmap, which will essentially put each variable side by side and tell us how strong of a correlation exists between two variables within the data set.

```python
# Step 1: Select key numeric variables (exclude non-numeric or categorical variables)
numeric_variables = ['Before-ClassworkStress', 'Now-ClassworkStress', 'FamilyRelationships', 'FriendRelationships']

# Step 2: Filter the dataframe to include only numeric columns
numeric_data = data[numeric_variables]

# Step 3: Calculate correlation matrix
correlation_matrix = numeric_data.corr()

# Visualize correlation matrix as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

Finally, let's examine the trends over time before and after the pandemic across our variables. We begin with Classwork Stress over time.

```python
# Trend analysis for Before-ClassworkStress and Now-ClassworkStress
plt.figure(figsize=(10, 6))

# Plot trend for Before-ClassworkStress
plt.plot(data.index, data['Before-ClassworkStress'], marker='o', label='Before-ClassworkStress')

# Plot trend for Now-ClassworkStress
plt.plot(data.index, data['Now-ClassworkStress'], marker='o', label='Now-ClassworkStress')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Stress Level')
plt.title('Trend Analysis of Classwork Stress')
plt.legend()

# Show plot
plt.show()
```

Next, we look at Homework Stress over time.

```python
# Group data by time period and calculate mean homework stress
homework_stress_over_time = data.groupby('Category')[['Before-HomeworkStress', 'Now-HomeworkStress']].mean()

# Plot line chart
plt.figure(figsize=(10, 6))
homework_stress_over_time.plot(marker='o')
plt.title('Changes in Homework Stress Over Time')
plt.xlabel('Time Period')
plt.ylabel('Homework Stress')
plt.xticks(range(2), ['Before', 'Now'])
plt.grid(True)
plt.legend(title='Time Period')
plt.show()
```

```python
# Group data by time period and calculate mean family and friend relationships
relationship_over_time = data.groupby('Category')[['FamilyRelationships', 'FriendRelationships']].mean()

# Plot line charts
plt.figure(figsize=(10, 6))

# Family Relationships over Time
plt.subplot(2, 1, 1)
plt.plot(relationship_over_time.index, relationship_over_time['FamilyRelationships'], marker='o', linestyle='-')
plt.title('Trends in Family Relationships Over Time')
plt.xlabel('Time Period')
plt.ylabel('Family Relationships (Mean)')

# Friend Relationships over Time
plt.subplot(2, 1, 2)
plt.plot(relationship_over_time.index, relationship_over_time['FriendRelationships'], marker='o', linestyle='-')
plt.title('Trends in Friend Relationships Over Time')
plt.xlabel('Time Period')
plt.ylabel('Friend Relationships (Mean)')

plt.tight_layout()
plt.show()
```

Finally, let's look at stress levels by state over time.

```python
# Group data by location and calculate mean stress levels
stress_by_location_time = data.groupby(['Country', 'State'])[['Before-ClassworkStress', 'Now-ClassworkStress']].mean()

# Plot line charts for stress levels by location over time
plt.figure(figsize=(12, 6))
for location, stress_data in stress_by_location_time.groupby(level=[0, 1]):
    plt.plot(stress_data.index.get_level_values('Country'), stress_data['Before-ClassworkStress'], marker='o', linestyle='-', label=f'Before - {location}')
    plt.plot(stress_data.index.get_level_values('Country'), stress_data['Now-ClassworkStress'], marker='o', linestyle='-', label=f'Now - {location}')
plt.title('Changes in Stress Levels by Location Over Time')
plt.xlabel('Country')
plt.ylabel('Stress Level')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Having concluded the EDA portion of our Analysis, here are some of our key findings that will guide us into the next phase of Preprocessing and training:

Age Distribution: The age distribution of students in the dataset spans a range of ages, with most falling between 13 and 18 years old.

Stress Levels: Stress levels, both before and after the pandemic, vary widely among students. Classwork stress appears to have increased slightly on average after the pandemic, while homework stress levels remained relatively stable.

Gender Differences: There are observable differences in stress levels between genders. Before and after the pandemic, females generally report higher stress levels compared to males.

Relationships: Family and friend relationships seem to have been impacted by the pandemic, with some students reporting decreased quality in these relationships over time.

Environmental Factors: The type of learning environment (physical vs. virtual) may influence stress levels, with variations observed in stress levels before and after the pandemic.
