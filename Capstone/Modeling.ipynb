```python
# Sample data for illustration purposes, replace this with your actual dataframe loading step
data = {
    'Author': ['A', 'B', 'C'],
    'Group': ['G1', 'G2', 'G3'],
    'T0': [1.0, 2.0, None],
    'Scale': ['Scale1', 'Scale2', 'Scale3'],
    'SD': ['SD1', 'SD2', 'SD3'],
    'N': [10.0, 20.0, 30.0],
    'T1': [1.1, None, 3.3],
    'Scale.1': ['Scale1', 'Scale2', 'Scale3'],
    'Measure.1': ['Measure1', 'Measure2', 'Measure3'],
    'SD.1': ['SD1', 'SD2', 'SD3'],
    'N.1': [11.0, 21.0, 31.0],
    'T2': [1.2, 2.2, 3.3],
    'Scale.2': ['Scale1', 'Scale2', 'Scale3'],
    'Measure.2': [1.5, 2.5, 3.5],
    'SD.2': [0.5, 0.6, 0.7],
    'N.2': [12.0, 22.0, 32.0]
}
interventions = pd.DataFrame(data)

# Define feature and target variables
X = interventions.drop(columns=['Author', 'Group'])
y = interventions['Group']

# Preprocessing for numeric data
numeric_features = ['T0', 'N', 'T1', 'N.1', 'T2', 'Measure.2', 'SD.2', 'N.2']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Preprocessing for categorical data
categorical_features = [col for col in X.columns if col not in numeric_features]
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# Define oversampling technique
oversampler = SMOTE(random_state=42)

# Define the classifier and parameter grid for grid search
clf_info = {
    'model': DecisionTreeClassifier(random_state=42),
    'params': {
        'classifier__max_depth': [3, 5, 10],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 4]
    }
}

# Create the pipeline
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('oversampler', oversampler),
    ('classifier', clf_info['model'])
])

# Use LeaveOneOut cross-validation
cv = LeaveOneOut()

# Perform grid search
grid_search = GridSearchCV(pipeline, clf_info['params'], cv=cv, n_jobs=-1, verbose=2)
grid_search.fit(X, y)

# Evaluate the model
# Since the dataset is smaller, we'll use the same data for evaluation
y_pred = grid_search.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Model: DecisionTree, Best Parameters: {grid_search.best_params_}, Accuracy: {accuracy}")

# Detailed classification report
print(classification_report(y, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=y.unique(), yticklabels=y.unique())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Save the model
joblib.dump(grid_search.best_estimator_, 'best_model.pkl')
print("Model saved as 'best_model.pkl'")

# Example of loading the model for future use
# loaded_model = joblib.load('best_model.pkl')
# loaded_model.predict(X)
```
